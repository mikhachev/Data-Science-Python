# Временные ряды 1.  Введение в задачу анализа временных рядов

## Введение

Временным рядом принято называть последовательно измеренные через некоторые (зачастую равные) промежутки времени данные [https://en.wikipedia.org/wiki/Time_series].
Посмотрим на примерах, что такое временной ряд и зачем его изучать.  

### Задача 1. Показать ситуацию

Оценка ситуации в момент, когда в этом возникла необходимость предполагает, что мы должны сделать вывод о ее качестве. Например, мы получили заработную плату за месяц в объеме 100000 р. Это хорошо, но непонятно стало лучше или хуже чем в прошлом. Другое дело, когда мы видим последовательные выплаты за год (5.01 - 100К₽, 5.02 - 100К₽, 5.03 - 100К₽ , …, 5.09 - 100К₽ ) . Стабильность - это приятно. Оценка ситуации возможно только с изучением предыстории. 

### Задача 2.  Предсказать, что будет
Предположим, что, изучив свой доход,  мы планируем когда-нибудь купить Кота. Конечно, правильной породы с родословной и привычкой к качественному корму. Вопрос стоит о том, когда следует купить котенка с точки зрения наилучшей цены. Очевидно, что стоимость котенка зависит от многих факторов и изменяется во времени по мере взросления котенка. Первые несколько месяцев она будет расти с котенком, но в какой-то момент наступит перелом (заводчику кошек - стало ясно, что продать котят не удастся) и начнется падение цены до уровня гордости заводчика. Конечно можно подождать по самого конца процесса, но котенок вырастет и не будет нас любить. Анализ рынка котят позволит выделить период начала падения, поможет определить как долго будет падать цена до стабилизации и определить приемлемое время покупки. А пока мы собираем денежки на кошку-нашей-мечты, а котенок немного подрастет. Правильная оценка истории подскажет, что будет впереди.

### Задача 3. Сопоставление факторов влияющих на целевой показатель

Пусть наша цель состоит в повышении уровня вашей заработной платы (мы завели котенка), но начальник не предлагает повышение, считая, что и так все прекрасно (у нас есть хлеб с маслом, а наш кот прекрасно сам может ловить мышей). Что сделает оптимист напишет служебную записку с требованием повысить зарплату (коту нужны дорогие консервы), пессимист будет экономить на масле, а мы, реалисты, исследуем состояние рынка предложений по уровню оплаты труда позиции data Science в hh (изменение окладов от времени) и график изменения популяции мышей в нашем доме (изменение популяции мышей от времени при появлении кота - внутренний скрытый фактор ) и напишем аналитический отчет об изменении во времени спроса и предложения на рынке в нашей профессиональной области, сопоставив его с ростом нашей производительности вследствие снятия стресса с помощью кота. Исследование различных факторов в исторической ретроспективе показывает их влияние друг на друга.

Таким образом, рассматривая явления во времени, обычно можно сказать о них значительно больше.
В этом цикле мы будем рассматривать различные методы обработки временных рядов направленные на их исследование для оценки и предсказания.

### Основные этапы обработки временных рядов разделены на несколько уроков :
1. Разведочный анализ.
2. Преобразование признаков для моделей прогнозирования.
3. Селекция признаков применительно к задаче прогнозирования.
4. Построение моделей прогнозирования временного ряда.


### Области применения анализа временных рядов:
- Оценка временных рядов (обработка сигналов, управление, обработка сигналов в системах связи) - наша  задача 1 про котика.
- Предсказание следующих значений (статистический анализ в экономике, предсказание цены валюты, акций и т.п., сейсмология - анализ сейсмограмм и предсказание землетрясений, метеорология и т.п.) - наша задача 2.
- Извлечение зависимостей и распознавание ( поиск взаимосвязей отдельных временных явлений, поиск аномалий и детектирование событий) - наша задача 3.


### Особенности временных рядов

Рассмотрим основные моменты связанные с представление и анализом информации во временных рядах. 

Все данные упорядочены относительно моментов времени, и, значит могут содержать дополнительную информацию, которую мы постараемся извлечь.
Метки времени позволяют связывать имеющиеся данные с другими временными рядами, которые при формировании информации не учитывались и не передавались специалисту. Например данные об объемах продажи и информация о событиях календаря или представление о сезонах года.
В данных есть объективные закономерности, явления повторяются, даже если это сразу не заметно 

- Во временных рядах принято выделять : 
    - Тренд (в ретроспективе 100 лет заработная плата наемных работников растет )
    - Циклическая составляющая (цикл солнечной активности 11 лет)
    - Сезонность (в течении года цена на яблоки ниже всего летом)
    - Шум (цена продажи квартиры в одном районе при прочих равных свойствах может отличаться на 1-2 %)

## Обработка временных меток в Pandas

Временная индексация в большинстве случаев удобна и работает достаточно эффективно. Обычно даты и время используют в качестве индексов структуры DataFrame. Однако поскольку временные метки содержат много дополнительной информации, то имеет смысл построить признаки, которые будут работать с ними и сопровождают изменения нашего временного ряда.

Для устранения влияния времени на результат и факторы при изучении взаимосвязанных рядов часто включают время события t в качестве независимой переменной в модель регрессии, что позволяет зафиксировать воздействие фактора t.

Попробуем реализовать подключение времени как признака , используя методы Pandas, который
предоставляет классы для обработки времени и даты разными способами. Существует типы данных для работы с временными данными: datetime64 и timedelta64.

Попробуем создать новую разметку времени. Для этого используют **метод pd.date_range(...)**. Пример 

```python
my_data_time = pd.date_range('1/1/2019', periods=15, freq="D")
my_data_time
```

создали разметку по дням начиная с 2019-01-01 и определили 15 дней. Посмотрим на формат описания подробней (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html) **pandas.date_range(start = <None или дата начала как строка в формате D.M.YYYY>, end = <None или дата конца>, periods= <None или число отсчетов>, freq=<None или можно задать дробность отсчета >, ...)**, где freq принимает значения :
    - 'D' - по умолчанию - интервал равен 1 дню,
    - 'M' - один месяц,
    - 'Y' - один год,
    - 'H' - один час,
    - 'T' - одна минута,
    - 'S' - секунда и т.д.
Все варианты представления интервалов можно проверить в документации функции. Еще один пример с месяцем как интервал времени

```python
my_data_time_M = pd.date_range('1/1/2018', '1-1-2019', freq="M")
my_data_time_M
```

```python
my_pd = pd.DataFrame(index = my_data_time_M, data = {'my_data': np.linspace(1,10,my_data_time_M.shape[0])})
my_pd.head()
```

Календарь - это объект типа DatetimeIndex, который содержит список праздничных дней.

Воспользуемся доступным USFederalHolidayCalendar. Для работы с календарем удобно пользоваться механизмом определения интервалов времени по датам timedelta из  datetime.
```python
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
from datetime import timedelta
one_day = timedelta(days=1)
cal = calendar()
#инициализировали объект календарь
data['is_holiday'] = data.index.get_level_values('Date').isin([d.date() for d in cal.holidays()])
data['is_prev_holiday'] = data.index.get_level_values('Date').isin([( d.date() - one_day ) for d in cal.holidays()])
data['is_next_holiday'] = data.index.get_level_values('Date').isin([( d.date() + one_day ) for d in cal.holidays()])
data['is_holiday'] = data['is_holiday'].astype('uint8')
data['is_prev_holiday'] = data['is_prev_holiday'].astype('uint8')
data['is_next_holiday'] = data['is_next_holiday'].astype('uint8')
```

## Разделение данных на тестовые и тренировочные во временных рядах

В классификаторе правильная модель разделения должна обеспечить одинаковое распределение примеров в пространстве признаков в тестовой и тренировочной выборках. Во временных рядах это не имеет смысла, т.к. предсказывая следующее значение необходимо использовать уже известную информацию. Получить доступ к данным из будущего у нас нет возможности (если бы мы могли предсказать, сколько ест котенок, когда подрастает, мы бы сразу нашли вторую работу). Это приводит к тому, что тестовая выборка при построении модели во временных рядах идет за массивом данных для тренировки. Пример такого разделения приведен ниже   

```python
# сделаем функцию и 
def split_data( data, split_date ):
    return data.loc[data.index.get_level_values('Date') <= split_date].copy(), \
           data.loc[data.index.get_level_values('Date') >  split_date].copy()
```

```python
train, test = split_data( data, '01-06-2016')

plt.figure( figsize=( 15, 7 ))
plt.xlabel('Время')
plt.ylabel('Потребление энергии, МВт')
plt.plot(train.index.get_level_values('Date'),train['PJME_MW'], label='train data' )
plt.plot(test.index.get_level_values('Date'),test['PJME_MW'], label='test data')
plt.title( 'Тестовые и тренировочные данные' )
plt.legend()
plt.show()
```

На уровне визуального анализа модель на основе случайного леса и градиентного бустинга выглядят более интересными, линейная регрессия не показывает длинные циклы годового изменения потребления. Оценим в цифрах промах моделей

**Средне Квадратичное Отклонение - Mean Square Error**

```python
er_l = mean_squared_error(y_true=y_test, y_pred=X_test_pred_l) 
er_f = mean_squared_error(y_true=y_test, y_pred=X_test_pred_f) 
er_g = mean_squared_error(y_true=y_test, y_pred=X_test_pred_g)

print('ошибка линейной регрессии :  ', er_l,'\nошибка случайного леса  :  ', er_f,'\nошибка градиентного бустинга :  ', er_g)
```

Очевидно, что градиентный бустинг победил со значительным отрывом. Однако можно проверить другую оценку, которая соотносит размер промаха с абсолютным значением целевого сигнала. Это средняя абсолютная процентная ошибка (MAPE)

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python

er_l = mean_absolute_percentage_error(y_true=y_test, y_pred=X_test_pred_l) 
er_f = mean_absolute_percentage_error(y_true=y_test, y_pred=X_test_pred_f) 
er_g = mean_absolute_percentage_error(y_true=y_test, y_pred=X_test_pred_g)

print('ошибка линейной регрессии :  ', er_l,'\nошибка случайного леса  :  ', er_f,'\nошибка градиентного бустинга :  ', er_g)
```

Отрыв градиентного бустинга сохраняется, но и случайный лес выглядит неплохо.
